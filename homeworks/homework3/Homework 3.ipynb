## The Background

There are three common means that can be associated with a group of values. You should already be familiar with the <b>arithmetic mean</b>, a.k.a. the average. The arithmetic mean is calculated by dividing the sum of all the values by the number of values, where each val is an individual value.

$$
arithmetic= \frac{\sum_{i=0}^{n-1}val_{i}}{n}
$$
    
For example, if we have 6 values of 3.2, 4.3, 1.7, 7.7, 5.2, and 8.1, the arithmetic mean would be

$$
geometric= \frac{{3.2+4.3+1.7+7.7+5.2+8.1}}{6}
$$

The second mean we'll work with is the <b>geometric mean</b>. It is calculated by taking the nth root of the product of all the values. The geometric mean is useful as a measure of central tendency, because it is less affected by extreme values than is the arithmetic mean. We find the geometric mean by multiplying all the values and then taking the nth-root, where n is the number of values:

$$
gemetric= \sqrt[n]{\prod_{i=0}^{n-1}val_{i}}
$$

For example, if we have 6 values of 3.2, 4.3, 1.7, 7.7, 5.2, and 8.1, the geometric mean would be:

$$
geometric= \sqrt[6]{3.2*4.3*1.7*7.7*5.2*8.1}
$$

The last mean is known as the <b>harmonic mean</b>. The harmonic mean is defined as the reciprocal of the arithmetic mean of the reciprocals. The harmonic mean is given by the formula

$$
harmonic=\frac{n}{\sum_{i=1}^{n}\frac{1}{val_{i}}}
$$

For example, if we have 6 values of 3.2, 4.3, 1.7, 7.7, 5.2, and 8.1, the harmonic mean would be:

$$
harmonic=\frac{6}{\frac{1}{3.2}+\frac{1}{4.3}+\frac{1}{1.7}+\frac{1}{7.7}+\frac{1}{5.2}+\frac{1}{8.1}}
$$
