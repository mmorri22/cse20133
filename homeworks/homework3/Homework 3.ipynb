    #include <stdlib.h>

    #define ROW 1
    #define COL 4

    void runArray( int prob_array[ROW][COL] );

    {
       int prob_array[ROW][COL];

       runArray( prob_array );

       runArray( prob_array );

       return 0;

    void runArray( int prob_array[ROW][COL] ){

        unsigned int iter;
        for( iter = 0; iter < ROW; iter++ ){

            unsigned int jter;
            for( jter = 0; jter < COL; jter++ ){

                prob_array[iter][jter] += iter+jter;

                fprintf( stdout, "%u ", prob_array[iter][jter] );
            }
            fprintf( stdout, "\n");
        }
