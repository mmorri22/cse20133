## Programming

Assignment Description: Given a phrase, you are to determine if any permutation of the phrase (ignoring whitespace and punctuation) could form a palindrome, which is a phrase that reads the same forwards and backwards. 

> For instance, "taco cat" is an example of a palindrome. And if taco cat is rearranged in any order, such as taccoat or aaccott are also valid, since they can be rearranged to be tacocat.



>  Inspiration: Note, this problem is inspired by <a href = "https://www.careercup.com/book">Problem 1.4</a> from Cracking the Code Interview and <a href = "https://www.interviewcake.com/question/python/permutation-palindrome">Question 30</a> from <a href = "https://www.interviewcake.com/">Interview Cake</a>.

For each input phrase, output a statement to an output file (which you may hard code the name)

"phrase" is a palindrome permutation

If the input phrase is a palindrome permutation. Otherwise, output:

"phrase" is a not palindrome permutation

### Example

Input:

    civic
    ivicc
    civil
    livci
    saippuakivikauppias

Output:

    "civic" is a palindrome permutation
    "ivicc" is a palindrome permutation
    "civil" is not a palindrome permutation
    "livci" is not a palindrome permutation
    "saippuakivikauppias" is a palindromic permutation
    
Before you work through the assignment, it is a good idea to draw out and figure out why ivicc works but livci does not work. This will help you narrow down a path forward.

Also, the palindromic permutation must be by character. For example, tacocaT is not a palindromic permutation.

> Hint: This requirement is designed to make the assignment simpler. Just focus on comparing the characters.
