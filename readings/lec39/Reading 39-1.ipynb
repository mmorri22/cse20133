## Algorithmic: Bridges of Koningsberg Problem

In 1786, Leohnard Euler was commissioned to solve the “Seven Bridges of Konigsberg” problem

<b>Problem:</b> - Does there exist a route that allowed a person to cross all seven bridges in Konigsberg without crossing any bridge twice?

    To this point, Everyone else is trying the Greedy approach
    The image below is a Google Maps drawing of the 7 Bridges is shown below



Euler realized that where a person starts is <i>irrelevant to the solution</i>! He translated the islands and bridges into locations

    Locations are Nodes (in graph theory, nodes are more commonly referred to a Vertices)
    Bridges are Edges
    
Then he combined the different river banks into one node each:



Euler also realized that, to get to an island, you can do one of three things:
<ol>
    <li>Leave and never come back</li>
    <li>Reach the island and leave the island</li>
    <li>Arrive to the island, which is where you finish</li>
</ol>


Therefore, he realized that if two nodes have an odd number of bridges and the rest have an even number, then the problem is solvable. 

Since more than two nodes have an odd number of bridges, <b>the problem is not solvable!</b>
