## Shortest Path Algorithm

> Also known as Dijkstra's algorithm for its inventor, <a href = "https://en.wikipedia.org/wiki/Edsger_W._Dijkstra">Edsger Dijkstra</a>.

Consider the following graph where there are several paths from the node containing 4 to the node containing 18.
<ul>
    <li>Each path has an <b>edge weight</b></li>
    <li>Goal: Find the path with the smallest sum of edge weights</li>
</ul>



    The following path: 4->20->10->18 encounters edges of weights 3 + 4 + 2 = 9

    Meanwhile, the path 4->20->30->18 encounters edges of weights 3 + 2 + 2 = 7

    Also, the path 4->5->10->18 encounters edges of weights 1 + 5 + 2 = 8

In a simple graph like this, we can tell with our own eyes which is the shortest path. But what if we are trying to determine the fewest number of satellites to transmit a signal? <b>We need a better way!</b>



## Approach 

Intuitively: The shortest path between u and v is measured by distance d

    If d(u,v) < d(u,x) + d(x,v)
        Then Edge (u,v) is the shortest path between them

Therefore, the shortest path for a graph contains the shortest path for the subgraphs that contain that path. Consider the diagram below.

I claim the path <b>4->5->10->18</b> is better than <b>4->20->10->18</b>

    Both paths contain the edge between the vertices containing 10->18, with a weight of 2
    Therefore, the shortest path to 10 is now the shortest path between these two paths
    
        Claim: <b>4->5->10</b> is better than <b>4->20->10</b>
    
    Now we compare the two paths to the vertex containing 10:

        4->5->10 < 4->20->10
        Weights: 3 + 4 < 1 + 5
        Weights: 7 < 6

### Setting up the algorithm

Initialize a Table for all nodes:

    Distance - Weight to destination - Initially from all but origin is infinity
    Parent - The previous node for the shortest path traversal - Initially -1 for all vertices
    Create a Priority Queue to evaluate the next values

We put them in when visited:

    Put on the PQ in Edge Weight Order

If dist[u] + weight < dist [v] of the current vertices we are evaluating

    Update dist[v] to dist[u] + weight
