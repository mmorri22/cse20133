## Building a Point

We will work together to build a class that represents a cartesian point on a plane

We will modify the <code>point.cpp</code> and <code>point.h</code> files.

In the .cpp file, you will create the method definitions

The .h file contained the method declarations

### Example

You need to include the class name in the function definition
Format -> return_type class_name::method_name(variables)

COORDINATE point::get_x()

void point::set_x(COORDINATE inX)

## Class Constructor

### Constructor Syntax
<ul>
    <li>A constructor has the same name as the class.</li>
    <li>A constructor has no return type (not even void).</li>
</ul>

### Default Constructors
<ul>
    <li>If you do not provide a constructor, the compiler provides a default constructor.</li>
    <li>The compiler-supplied default constructor initializes each data member to its default value.</li>
</ul>

### In the Header File - Constructor Format

		point();
		
		point( const COORDINATE& x_in, const COORDINATE& y_in );
    
    
### Assignment Inputs

PQC rules state that we must input them in the order the are declared in the class

    point::point() : x_coor(0), y_coor(0) {}


    point::point( const COORDINATE& x_in, const COORDINATE& y_in )
        : x_coor(x_in), y_coor(y_in) {}
        
        
        
## We will build class method examples


		COORDINATE get_x() const;

		COORDINATE get_y() const;
		
		void set_y( const COORDINATE& y_in );
		
		void print() const;
        
        
## And here are the methods in the .cpp file

    COORDINATE point::get_x() const{

        return x_coor;
    }

    COORDINATE point::get_y() const{

        return y_coor;
    }

    void point::print() const{

        COUT << "{" << x_coor << ", " << y_coor << "}";
    }

    void point::set_y( const COORDINATE& y_in ){

        y_coor = y_in;
    }
