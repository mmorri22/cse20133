## In-Class Lecture 36-3 - Electric Field Visualization Example

In this in-class example, we will use Matplot++ to visualize an Electric Field. We will continue this into Lecture 37 if necessary.

### Problem Example

Consider a point charge $Q=1nC$ located at $(x=1,y=1,z=4)$. The vector electric field at a point P from the point charge is given by:<p>

<center>$\bar{E}=\frac{CQ}{R^{2}}\hat{R}$</center><p>
    
where $C=9x10^{9}$, $R$ is the separation distance from the point charge to point $P$, and $\hat{R}$ is a unit vector point from the point charge to point $P$.
    
We will generate a three-dimensional vector plot of the electric field associated with the change over the range $0<x<2$, $0<y<2$, $3<z<5$. Note that the charge exists outside the plotted region.

#### Part 1

Open <code>electric.cpp</code>

Create a linspace for the ranges of x_range, y_range, and z_range with 10 values in each element (+- 1 on eacg side

Then, we will create six vectors:
<ul>
    <li>A vector to represent the final x locations called x_coor</li>
    <li>A vector to represent the final y locations called y_coor</li>
    <li>A vector to represent the final z locations called z_coor</li>
    <li>A vector to represent the final Ex electric field called electric_x</li>
    <li>A vector to represent the final Ey electric field called electric_y</li>
    <li>A vector to represent the final Ez electric field called electric_z</li>
</ul>

#### Part 2

We will loop through each of the x, y, and z ranges

> This is $O(N^{3})$ run time!

Copy the current value from the array to local copies (this will improve run time as we have observed in experiment)

Next, we will push these local values on the back of coor vectors. Now we will have every possible element in 3 dimensions

Calculate the diff we will need for the vector for x, y, and z

We will next use <code>pow</code> to calculate the unit vector $\hat{R}$ 

Then, we will calculate the Electric Field $\bar{E}$ for Ex, Ey, and Ez

Finally, we will push those values onto the back to electric_x, electric_y, and electric_z.

#### Part 3

We will use quiver3 to represent pointing and vectors

    quiver3(x_coor, y_coor, z_coor, electric_x, electric_y, electric_z);

    show();
