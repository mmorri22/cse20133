    void print_func(int the_array_pf[], long unsigned int* array_len_pf ){

        fprintf( stdout, "%p %lu %p %p\n", the_array_pf, *array_len_pf, array_len_pf, &array_len_pf);

        for( iter = 0; iter < *array_len_pf; ++iter ){
